### Exercício 10

```python
# Crie um banco de dados no SQLite, crie uma tabela, insira registros, 
# consulte a tabela e retorne os dados em dataframe do Pandas

import sqlite3
import pandas as pd
query = """
CREATE TABLE GITHUB
(Number INTEGER NOT NULL,
Title VARCHAR(150),
Labels VARCHAR(255),
State VARCHAR(20)
);"""

con = sqlite3.connect('issues_github.db')
con.execute(query)
con.commit()
```

```python
from sqlalchemy import create_engine

engine = create_engine('sqlite:///issues_github.db', echo=True)
sqlite_connection = engine.connect()
```

```python
sqlite_table = 'GITHUB'
issues_save = pd.DataFrame(data, columns=['number', 'title', 'state'])
issues_save.head()
issues_save.to_sql(sqlite_table, sqlite_connection, if_exists='replace')
```

```python
cursor = con.execute('SELECT * FROM GITHUB')
rows = cursor.fetchall()
rows
```

```python
pd.DataFrame(rows, columns=[x[0] for x in cursor.description])
```
