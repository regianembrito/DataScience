### Graficos de Barras

```python
x = [2, 4, 6, 8, 10]
y = [6, 7, 8, 2, 4]
```

```python
plt.bar(x, y, label = 'Barras', color='r')
plt.legend()
plt.show()
```

```python
x2 = [1, 3, 5, 7, 9]
y2 = [7, 8, 2, 4, 2]
```

```python
plt.bar(x, y, label='Barras1', color='r')
plt.bar(x2, y2, label='Barras2', color='y')
plt.legend()
plt.show()
```

```python
idades = [22, 65, 45, 55, 21, 22, 34, 42, 41, 4, 99, 101, 120, 122, 130, 111, 115, 80, 75, 54, 44, 64, 13, 18, 48]
ids = [x for x in range(len(idades))]
plt.bar(ids, idades)
plt.show()
```

```python
bins = list(range(0, 140, 10))
bins
```

```python
plt.hist(idades, bins, histtype='bar', rwidth=0.8)
plt.show()
```

```python
plt.hist(idades, bins, histtype='stepfilled', rwidth=0.8)
plt.show()
```

### Scatterplot

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,5,6,8,4,8]
```

```python
plt.scatter(x,y, label='Pontos', color='r', marker='o', s=100)
plt.legend()
plt.show()
```

### Stack Plots

```python
dias = [1,2,3,4,5]
dormir = [7,8,6,77,7]
comer = [2,3,4,5,3]
trabalhar = [7,8,7,2,2]
passear = [8,5,7,8,13]
```

```python
plt.stackplot(dias, dormir, comer, trabalhar, passear, colors=['m', 'c', 'r', 'k', 'b'])
plt.show()
```

### Pie Chart

```python
fatias = [7, 2, 2, 13]
atividades = ['dormir', 'comer', 'trabalhar', 'passear']
colunas = ['c', 'm', 'r', 'k']
```

```python
plt.pie(fatias, labels=atividades, colors=colunas, startangle=90, shadow=True, explode=(0,0.1,0,0))
plt.show()
```

### Pylab

```python
# Visualizando os gráficos dentro do Jupyter Notebook
# O Pylab combina funcionalidades do pyplot com as do Numpy
from pylab import *
%matplotlib inline
```

```python
x = linspace(0,5,10)
y = x**2
fig = plt.figure()

# Definindo os eixos
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Gráfico de Linha')
```

```python
# Gráficos com 2 figuras
x = linspace(0, 5, 10)
y = x ** 2

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # eixos da figura principal
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # eixos da figura secundária

# Figura principal
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Figura Principal')

# Figura secundária
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Figura Secundária')
```

```python
# Gráficos em Paralelo
fig, axes = plt.subplots(nrows=1,ncols=2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Título')
    
fig.tight_layout()
```
