```python
def tratar_nome(linha):
    match = re.compile(r'"(.*)(,)(\s.*)"')
    linha_tratada = match.sub(r'"\3 \1"', linha)
    return linha_tratada.split(",")

lista = [];
#lista = lambda lista: tratar_nome(linha)


dados_tratados = [tratar_nome(linha) for linha in linhas[1:]]
dados_tratados
```

```python
titanic_dados = {}
for coluna in colunas:
    titanic_dados[coluna] = []
    
titanic_dados
```

```python
for passageiro in dados_tratados:
    for indice, dado in enumerate(passageiro):
        coluna = colunas[indice]
        titanic_dados[coluna].append(dado)
titanic_dados["Age"]
#int(titanic_dados["Age"][0])
```

```python
ages = []
for indice, age in enumerate(titanic_dados["Age"]):
    if titanic_dados['Age'][indice] != '':
        titanic_dados['Age'][indice] = float(age)
        ages.append(float(age))
    else:
        titanic_dados['Age'][indice] = -1

type(titanic_dados['Age'][0])
```

```python
lista_idades = [age for age in titanic_dados['Age'] if age != -1]
```

```python
print(lista_idades)
```

```python
def media(lista):
    total = sum(lista)
    quantidade = len(lista)
    return total / quantidade

print( round(media(lista_idades), 2))
```
